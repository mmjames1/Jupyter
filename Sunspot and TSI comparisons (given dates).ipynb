```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

The first set of graphs here shows the data for the sunspot count (yellow) and the change of the TSI (blue) for the period between 8/26/2017 and 9/26/2017

```python
#sunspot_tbl = ascii.read("SS_sept2017.txt")
#TSI_tbl = ascii.read("TSI_sept2017.txt")

df_SS18 = pd.read_csv('/Users/mmjames1/Data and Code/SS_feb2018.csv')
df_SS17 = pd.read_csv('/Users/mmjames1/Data and Code/SS_sept2017.csv')
df_TSI18 = pd.read_csv('/Users/mmjames1/Data and Code/TSI_feb2018.csv')
df_TSI17 = pd.read_csv('/Users/mmjames1/Data and Code/TSI_sept2017.csv')
```

Subsetting the columns to make lists for plots

```python
#feb 2018 date
date_18_df = df_SS18['Date']
SS_18_df = df_SS18['Total']
TSI_18_df = df_TSI18['TSI']

#sept 2017 date
date_17_df = df_SS17['Date']
SS_17_df = df_SS17['Total']
TSI_17_df = df_TSI17['TSI']
```

```python
#making the lists from the csv subsets
date_18 = date_18_df.tolist()
SS_18 = SS_18_df.tolist()
TSI_18 = TSI_18_df.tolist()

date_17 = date_17_df.tolist()
SS_17 = SS_17_df.tolist()
TSI_17 = TSI_17_df.tolist()
```

### The plots for Feb 2018

```python
f, axarr = plt.subplots(2, sharex=True)

axarr[0].plot(date_18,SS_18,'y')
axarr[0].set_ylabel('Sunspot Counts')
axarr[0].grid(True)

axarr[1].plot(date_18,TSI_18,'b')
axarr[1].set_xlabel('Date')
axarr[1].set_ylabel('TSI Changes')
axarr[1].grid(True)
```

```python
max_ss_18 = max(SS_18)
min_TSI_18 = min(TSI_18)
print(max_ss_18,min_TSI_18)
```

```python
print(SS_18.index(max_ss_18))
print(TSI_18.index(min_TSI_18))
```

the min TSI and max SS number seem to be at the same time (according to the index) but the decrease in the TSI doesn't appear to be the same width as the increase in SS time is
